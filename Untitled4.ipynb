```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('IPL.csv')
```

```python
# which team won the most matches?

match_wins = df['match_winner'].value_counts()

sns.barplot(x = match_wins.index, y = match_wins.values, palette = 'viridis')
plt.xlabel('Team')
plt.ylabel('Number of Matches Won')
plt.title('Matches Won by Each Team')
plt.xticks(rotation = 90)
plt.show()
```

```python
df.head()
```

```python
# Toss decision trends
df['toss_decision'].value_counts()
```

```python
sns.countplot(x=df['toss_decision'], palette='rainbow')
plt.title('Toss Decision Distribution')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.show()
```

```python
# Toss winner Vs Match winner
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count * 100) / df.shape[0]
print('Winning percentage who win the toss and match :',percentage)
```

```python
df
```

```python
# How do teams win? (Runs vs Wickets)
sns.countplot(x=df['won_by'], palette='rainbow')
plt.title('Winning by Runs or Wickets')
plt.xlabel('Decision')
```

```python
# Most "Player of the Match" Awards
count = df['player_of_the_match'].value_counts().head(10)
print(count)
```

```python
sns.barplot(x=count.index, y=count.values, palette='rainbow')
plt.title('Most Player of the Match Awards')
plt.xlabel('Players')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top Scorers
count = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
count
```

```python
sns.barplot(x=count.index, y=count.values, palette='viridis')
plt.title('Top Scorers')
plt.xlabel('Players')
plt.ylabel('Runs')

plt.show()
```

```python
# Most Matches Played by Venue
count = df['venue'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.index, y=count.values, palette='rainbow')
plt.title('Most Matches Played by Venue')
plt.xlabel('Venue')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Which player had the highest individual score?
df[df['highscore'] == df['highscore'].max()][['top_scorer', 'highscore']]
```

```python
# 10 Best Bowling Figures
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='bar')
```

```python
df
```

```python
# Which bowler had the best bowling figures?
df[df['highest_wickets'] == df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```

```python
# Who won the highest margin by runs?
df[df['won_by'] == 'Runs'].sort_values(by='margin', ascending=False).head(1)[['match_winner', 'margin']]
```
